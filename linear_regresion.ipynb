```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
dataset=pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
```

```python
x=dataset.iloc[:,1:(len(dataset.columns)-1)].values
```

```python
y=dataset.iloc[:,(len(dataset.columns)-1)].values
```

```python
plt.plot(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm=lm.fit(x_train,y_train)
```

```python
print(lm)
```

```python
sam=lm.predict(x_test)
```

```python
plt.subplot(211)
plt.plot(y_test,color='g',label='actual')
#plt.plot(sam,color='c',label='predicted')
plt.legend()

plt.subplot(212)
plt.plot(sam,color='c',label='predicted')
plt.legend()



```

```python
plt.plot(y_test,color='g',label='actual')
plt.plot(sam,color='r',label='predicted')
plt.legend()
```
